```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
path = 'C:/windows/fonts/H2GTRM.TTF'
font_name = fm.FontProperties(fname = path).get_name()
plt.rc('font',family=font_name)
f = open('data/가족회사운영현황.csv')
data = csv.reader(f)

header = next(data)
data1= []
xlabel= []
next(data)
# print(header)
xlabel.append(header[4])
xlabel.append(header[7])
xlabel.append(header[10])
xlabel.append(header[13])
xlabel.append(header[16])
xlabel.append(header[19])

# print(xlabel)
data2=[]
for row in data:
#     전체 데이터 수집
    if row[0] =="전체":
        row[3:]=map(int,row[3:]);
        if row[2] != "소계":

            d= row[4],row[7],row[10],row[13],row[16],row[19]
            data2.append(row[2])
            data1.append(d)


            


# 그래프화
plt.figure(figsize=(20,10))
for i in range(len(data2)):
    plt.plot(xlabel,data1[i], label=data2[i])
plt.legend()
plt.show()


# 막대그래프
for i in range(len(data2)):
    plt.figure(figsize = (20,10))
    plt.title(data2[i])
    plt.bar(xlabel,data1[i])
    plt.show()

# plt.figure(figsize = (20,10))
# plt.title(data2[0])
# plt.bar(xlabel,data1[0])
# plt.show()
        
```
